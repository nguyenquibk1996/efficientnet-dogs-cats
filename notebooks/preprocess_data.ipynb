```python
from pathlib import Path

import os
import pandas as pd
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
```

```python
IMAGE_LIST_PATH = '../data/train'

IMAGE_SAVE_PATH = '../data/annotations/all_list.csv'
TRAIN_SAVE_PATH = '../data/annotations/train.csv'
VAL_SAVE_PATH = '../data/annotations/val.csv'
```

```python
image_list = os.listdir(IMAGE_LIST_PATH)
```

```python
image_list
```

```python
labels = []
for image in image_list:
    label = image[:3]
    if label == 'cat':
        label = 0
    else:
        label = 1
    labels.append(label)
```

```python
labels
```

```python
all_image_with_label_df = pd.DataFrame({
    'image_name': image_list,
    'label': labels
})
```

```python
all_image_with_label_df
```

```python
all_image_with_label_df.to_csv(IMAGE_SAVE_PATH)
```
